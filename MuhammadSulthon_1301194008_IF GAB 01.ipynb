Proses Filter LPF

```python
fs_rate = len(df_can['Linear Acceleration z (m/s^2)'])/60       
T = 60         
cutoff = 0.8      
nyq = 0.5 * fs_rate  
order = 5       
n = len(df_can['Linear Acceleration z (m/s^2)']) 

def butter_lowpass_filter(data, cutoff, fs_rate, order):
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y
```

plot data sinyal sebelum di filter

```python
y = butter_lowpass_filter(df_can['Linear Acceleration z (m/s^2)'], cutoff, fs_rate, order)
fig = go.Figure()
fig.add_trace(go.Scatter(
            y = df_can['Linear Acceleration z (m/s^2)'],
            x = df_can["Time (s)"],
            line =  dict(shape =  'spline' ),
            name = 'signal dengan noise'
            ))
fig.add_trace(go.Scatter(
            y = y,
            x = df_can["Time (s)"],
            line =  dict(shape =  'spline' ),
            name = 'signal hasil filter'
            ))
fig.show()
```

plot data sinyal sesudah dilakukan LPF

```python
y2 = butter_lowpass_filter(df_can['inklinasi'], cutoff, fs_rate, order)
fig2_plot = go.Figure()
fig2_plot.add_trace(go.Scatter(
            y = df_can['inklinasi'],
            x = df_can["Time (s)"],
            line =  dict(shape =  'spline' ),
            name = 'signal dengan noise'
            ))
fig2_plot.add_trace(go.Scatter(
            y = y2,
            x = df_can["Time (s)"],
            line =  dict(shape =  'spline' ),
            name = 'signal hasil filter'
            ))
fig2_plot.show()
```

```python
# dikarenakan data yang di olah merupakan  data digital maka digunakanlah fungsi butter untuk melakukan filter LPF
# frekuensi sinyal sebelum dilakukan filter hampir tidak terlihat karena berada pada kisaran frekuensi 0
# Keadaan setelah dilakukan filter LPF 
# pada metode direct amplitudo tertinggi yang sebelumnya berada pada 1,4 derajat menjadi sekitar 0,03 derajat
# pada metode inklinasi sudut tertinggi yang tadinya sekitar 180 derajat berubah menjadi kisaran 50 derajat
```
